NUMERICAL ANALYSIS AND DATA VISUALIZATION

```python
for col in ["Engine_Size","Price","Mileage"]:
    print("Skewness of "+col, df[col].skew() )
```

```python
for col in ["Engine_Size","Price","Mileage"]:
    mp.hist(df[col],bins=10,edgecolor='black',color='lavender')
    mp.title("Distribution of "+ col)
    mp.xlabel(col)
    mp.show()
    df[col].skew()
```

CATEGORICAL ANALYSIS AND DATA VISUALIZATION

```python
for col in ["Year","Doors","Fuel_Type","Transmission","Owner_Count"]:
    print(df[col].value_counts())
```

```python
for col in ["Year","Doors"]:
    print("Skewness of "+col, df[col].skew() )
```

```python
for col in ["Year","Doors","Fuel_Type","Transmission"]:
    sns.countplot(df, x=col,edgecolor='black',palette="Set2")
    mp.title("Distribution of "+ col)
    mp.xlabel(col)
    mp.xticks(rotation=90)
    mp.show()
```

```python
mp.figure(figsize=(10,6))
yearly_owners=df.groupby("Year")["Owner_Count"].sum().reset_index()
sns.barplot(x="Year",y="Owner_Count",data=yearly_owners,palette="coolwarm",hue="Owner_Count")
mp.xticks(rotation=90)
mp.title("Owner Count vs Year")
mp.show()
```

```python
mp.figure(figsize=(10,6))
brand_owners=df.groupby("Brand")["Owner_Count"].sum().reset_index()
sns.barplot(x="Brand",y="Owner_Count",data=brand_owners,palette="coolwarm",hue="Owner_Count")
mp.xticks(rotation=90)
mp.title("Owner Count vs Brand")
mp.show()
```
